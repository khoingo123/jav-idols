```
# Import necessary packages
from selenium import webdriver
import numpy as np
import pandas as pd
from google.colab import drive
```

```
# Scrape data from the website
JAV_idols = pd.DataFrame(columns = ['Name','Favorite', 'Date_of_birth', 'Blood_type', 'Hometown', 'Height', 'Size', 'Hobby', 'Special_skill'])
link_page = 'https://xxx.xcity.jp/idol/'
driver.get(link_page)       # get link

for i in range(7, 11): #10
    num_cols = len(driver.find_elements_by_xpath('//*[@id="leftCol"]/div[1]/div[1]/table/tbody/tr[{}]/td'.format(i)))
    for j in range(1, num_cols + 1): #num_cols + 1
        try:
            link_alphabet = driver.find_elements_by_xpath('//*[@id="leftCol"]/div[1]/div[1]/table/tbody/tr[{}]/td[{}]/a'.format(i, j))[0].get_attribute('href')
        except:
            continue
        driver.get(link_alphabet)       # get link

        num_pages = int(driver.find_elements_by_xpath('//*[@id="avidol"]/ul[1]/*')[-2].text)
        for k in range(1, num_pages + 1): #num_pages + 1
            link_number = link_alphabet + '&page={}'.format(k)
            driver.get(link_number)       # get link

            num_idols = len(driver.find_elements_by_xpath('//div[@class="itemBox"]'))
            for l in range(1, num_idols + 1):
                try:
                    link_idol = driver.find_elements_by_xpath('//*[@id="avidol"]/div[{}]/div/p[2]/a'.format(l))[0].get_attribute('href')
                except:
                    link_idol = driver.find_elements_by_xpath('//*[@id="avidol"]/div[{}]/div/p/a'.format(l))[0].get_attribute('href')
                
                driver.get(link_idol)       # get link

                list_info = driver.find_element_by_xpath('//dl[@class="profile"]').text.split('\n')
                if 'Favorite' not in list_info[1]:
                    list_info.insert(1, '')
                # Extract name
                name = driver.find_elements_by_xpath('//*[@id="avidolDetails"]/div/h1')[0].text

                # Extract favorite
                favorite = list_info[1][10:]

                # Extract date_of_birth
                date_of_birth = list_info[2][13:]

                # Extract blood_type
                blood_type = list_info[3][10:]

                # Extract hometown
                hometown = list_info[4][12:]

                # Extract height
                height = list_info[5][6:-2]

                # Extract size
                size = list_info[6][4:]

                # Extract hobby
                hobby = list_info[7][5:]

                # Extract special_skill
                special_skill = list_info[8][13:]

                JAV_idols.loc[len(JAV_idols)] = [name, favorite, date_of_birth, blood_type, hometown, height, size, hobby, special_skill]
                driver.get(link_number)       # get link
            
            driver.get(link_alphabet)       # get link
        
        driver.get(link_page)       # get link
```
