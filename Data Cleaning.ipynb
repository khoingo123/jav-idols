```
# Import necessary libraries
import os
import pandas as pd
import numpy as np
import math

files = os.listdir()
```

```
# Concatenate many single files into one
data = pd.read_csv(files[0])
for file in files[1:]:
    other_data = pd.read_csv(file)
    data = pd.concat([data, other_data], ignore_index=True)
```

## II. Clean data

```
data.head(20)
```

```
# Clean name, delete the [] part
def clean_name(name):
    if '[' in name:
        index = name.index('[')
        name = name[:index]
    return name
    
data['Name'] = data['Name'].apply(lambda x: clean_name(x))
```

```
# Clean favorite by converting to int16
data['Favorite'] = data['Favorite'].astype('Int16')
```

```
# Clean birth of date by converting to birth of year
data['Date_of_birth'] = data['Date_of_birth'].fillna(np.nan).str[:4].astype('float').astype('Int16')
data.rename(columns={"Date_of_birth": "Year_of_birth"}, inplace=True)
```

```
# Clean height by converting to int64
data['Height'] = data['Height'].astype('Int16')
```

```
data.head(20)
```

```
# Split the size to a list
for new_col in ['Breast','Waist','Hip']:
    data[new_col] = np.nan

data['Size'] = data['Size'].str.split(' ')
```

```
# remove parentheses in size
def remove_parentheses(list_size):
    try:
        for i in range(len(list_size)):
            if '(' in list_size[i]:
                idx = list_size[i].index('(')
                list_size[i] = list_size[i][:idx]
    except:
        pass
    return list_size

data['Size'] = data['Size'].apply(lambda x: remove_parentheses(x))
```

```
# Split size column to 3 separate columns
for i in range(len(data)):
    list_size = data.loc[i, 'Size']
    try:
        for size in list_size:
            if 'B' in size:
                data.loc[i, 'Breast'] = size[1:]
            if 'W' in size:
                data.loc[i, 'Waist'] = size[1:]
            if 'H' in size:
                data.loc[i, 'Hip'] = size[1:]
    except:
        pass
        
data.drop(columns=['Size'], inplace=True)
```

```
# Clean hometown
data['Hometown'] = data['Hometown'].str.lower()
for i in range(len(data)):
    try:
        data.loc[i, 'Hometown'] = dict_hometown[data.loc[i, 'Hometown']]
    except:
        pass
```

```
# Clean hobby and special skill
from googletrans import Translator
translator = Translator()

for i in range(len(data)):
    try:
        while translator.detect(data.loc[i, 'Hobby']).lang != 'en':
            data.loc[i, 'Hobby'] = translator.translate(data.loc[i, 'Hobby'], dest='en').text
        while translator.detect(data.loc[i, 'Special_skill']).lang != 'en':
            data.loc[i, 'Special_skill'] = translator.translate(data.loc[i, 'Special_skill'], dest='en').text
        print(i)
        print(data.loc[i, 'Hobby'])
        print(data.loc[i, 'Special_skill'])
    except:
        pass
```
